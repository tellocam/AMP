```python
maxThreads = 12
total_acqs = 1000
bench_iters = 100

df_LockOMP = pd.read_csv("benchLockOMP_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_CriticalOMP = pd.read_csv("benchCriticalOMP_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_TAS = pd.read_csv("benchTAS_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_TATAS = pd.read_csv("benchTATAS_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_Ticket = pd.read_csv("benchTicket_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
# df_array = pd.read_csv("benchLockOMP_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_CLH = pd.read_csv("benchCLH_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_MCS = pd.read_csv("benchMCS_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
df_Hemlock = pd.read_csv("benchHemlock_THR64_ACQ1000_ITER100_Nebula.txt", delimiter='\t').fillna(0)
```

```python
plotWhat = "Time"  # Time, Wait, Fail, Fair, TP - these strings can be used in the plotWhat variable plotWhat
LockList = ["OMP Lock", "OMP Critical", "TAS", "TATAS", "Ticket", "Array", "CLH", ]

# utils.plotStuff("threads", plotWhat, "OMP Lock", df_LockOMP, "black")
utils.plotStuff("threads", plotWhat, "OMP Critical", df_CriticalOMP, color = "red")
utils.plotStuff("threads", plotWhat, "TAS", df_TAS, "blue")
utils.plotStuff("threads", plotWhat, "TATAS", df_TATAS, "green")
utils.plotStuff("threads", plotWhat, "Ticket", df_Ticket, "magenta")
# utils.plotStuff("threads", plotWhat, "Array", df_array, "yellow")
utils.plotStuff("threads", plotWhat, "CLH", df_CLH, "violet")
utils.plotStuff("threads", plotWhat, "MCS", df_MCS, "gray")
utils.plotStuff("threads", plotWhat, "Hemlock", df_MCS, "orange")

plt.title("Avg. Total Time - 1000 iters. each with 1000 acq.")
plt.xlabel("Threads / -")
plt.ylabel("Avg. Total Time / s")
plt.legend()
plt.grid()
```
